```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("quadeer15sh/augmented-forest-segmentation")

print("Path to dataset files:", path)
```

```python
import matplotlib.pyplot as plt
import os
# Display some images from the dataset  

images_path = "dataset/Forest Segmented/Forest Segmented/images"
masks_path = "dataset/Forest Segmented/Forest Segmented/masks"
image_files = [f for f in os.listdir(images_path) if f.endswith('.jpg')]
mask_files = [f for f in os.listdir(masks_path) if f.endswith('.jpg')]   

# Ensure the number of images and masks match
if len(image_files) != len(mask_files):
    raise ValueError("The number of images and masks do not match.")

#plot image and mask pairs
num_images_to_display = 2 
plt.figure(figsize=(10,5))
if num_images_to_display > len(image_files):
    num_images_to_display = len(image_files)
    print(f"Only {num_images_to_display} images available to display.")

for i in range(num_images_to_display):
    img_path = os.path.join(images_path, image_files[i])
    mask_path = os.path.join(masks_path, mask_files[i])

    img = plt.imread(img_path)
    mask = plt.imread(mask_path)

    plt.subplot(num_images_to_display, 2, 2*i + 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Image {i+1}")

    plt.subplot(num_images_to_display, 2, 2*i + 2)
    plt.imshow(mask, cmap='gray')
    plt.axis('off')
    plt.title(f"Mask {i+1}")
```
